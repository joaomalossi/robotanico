```python
# Carregando bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# Carregando o csv em um dataframe
df = pd.read_csv('iris.csv')
```

```python
# Plotando a dispersÃ£o dos dados
sb.pairplot(df, hue='target')
```

```python
# Selecionando as features para classificaÃ§Ã£o em um array NUMPY
X = np.array(df.drop('target',1))
X
```

```python
# Selecionando as classes para classificaÃ§Ã£o em um array NUMPY
y = np.array(df.target)
y
```

```python
# Classificando com o KNN
# Importando o KNN do scikit-learn
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Criando um classificador
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Treinando o classificador
knn.fit(X,y)
```

```python
# Predizendo a flor iris
knn.predict([[6.5,6.5,4.7,1.3]])
```
